# Difracción

La difracción es el fenómeno que se produce cuando la luz, al pasar por aperturas pequeñas, o por obstáculos pequeños  (comparables a su longitud de onda), se desvía de su propagación rectilínea. La misma puede ser entendida a nivel fenomenológico utilizando el principio de Huygens-Fresnel, que establece que cada punto de un frente de onda, en un instante de tiempo dado, se comporta como una fuente de ondas (secundarias) esféricas:

<img src="./imgDiff/esfericas.jpg" alt="Drawing" style="width: 300px;"/>

La intensidad de la luz en cualquier punto adelante de la fuente es debida a la interferencia de todas estas ondas. Aunque cada punto individual genera una onda esférica, la interferencia de todas ellas genera una onda plana que viaja en la misma dirección que la onda inicial.

Cuando el frente de onda encuentra un obstáculo, algunos puntos fuente de ondas secundarias desaparecen y el resultado es un patrón de interferencia diferente a una onda plana, que se denomina **patrón de difracción**. En particular, los puntos que se encuentran en los extremos del frente de onda obstruido no tienen al lado otros puntos que generen ondas que interfieran con las ondas que ellos generan. Las ondas entonces generadas por estos puntos se aproximan a ondas esféricas o cilíndricas. Como consecuencia, al adoptar el frente de onda una forma redondeada en donde fue recortado, la dirección de propagación de la onda cambia, girando hacia el obstáculo.

# Difracción de Fraunhofer

Los fenómenos de difracción se clasifican en dos categorías, que dependen de la distancia entre el objeto difractante y el punto de observación. Cuando la distancia no es grande, la imagen en el punto de observación preserva similitud con la forma geométrica del objeto difractante (es decir, el obstáculo). 

Cuando la distancia es grande, la distribución de la densidad de energía resultante de la interferencia entre las ondas que se propagan después de la apertura es un patrón de franjas claras y oscuras. El patrón de difracción permite determinar las dimensiones del objecto que difracta. Este último es el **modelo de Fraunhofer**, y es **válido cuando la distancia entre el objeto difractante (apertura u obstaculo) y el punto de observación es mucho mayor que el tamaño de la apertura**. 

El modelo de Fraunhofer es muy utilizado porque además de proporcionar resultados correctos para la mayoría de las experiencias sencillas que se pueden realizar en el laboratorio, es de los que implican cálculos más sencillos. El otro modelo de difracción es el de Fresnel, o de campo próximo, válido cuando el punto de observación se encuentra cerca del objeto que difracta.

## Difracción por una rendija simple

Para estudiar la difracción generada por una apertura consideremos que la luz llega como una onda plana, perpendicular a la apertura, y que cada punto de la apertura se comporta como una fuente puntual. El patrón de difracción entonces se modela como el patrón debido a la interferencia de la radiación emitida por esas fuentes puntuales.

Una onda plana puede ser generada experimentalmente por un láser colimado, o por una lámpara de luz blanca. En este último caso hay que poner el punto de observación a una distancia grande del punto de emisión de la luz con el fin de obtener sólo rayos paraxiales (en la aproximación paraxial las trayectorias de los rayos de luz forman ángulos pequeños con el eje óptico, siendo el eje óptico perpendicular al plano de la imagen, pudiendose así aproximar el seno y la tangente de un ángulo por el ángulo mismo).

La intensidad de luz medida en la pantalla es igual al valor medio del cuadrado del campo eléctrico de la onda. Se demuestra que la expresión de la intensidad de luz (en función del ángulo $\theta$) vale: 
\begin{equation}
I(\theta)~=~I_0 sinc^2 \left(\frac{b \pi \tan (\theta)}{\lambda}\right)
\end{equation}

donde la función *sinc* (seno cardinal) se define como:
\begin{equation}
sinc(x)=\frac{\sin(x)}{x}
\end{equation}

El patrón de difracción presenta una serie de máximos y mínimos que dependen, además de la longitud de onda utilizada, del ancho de la rendija, como se puede observar a continuación (en la figura, *w* es el ancho de la rendija):
<img src="./imgDiff/oneslit.jpg" alt="Drawing" style="width: 300px;"/>

El siguiente script permite simular la intensidad observada, variando el ancho de la rendija, la longitud de onda y la distancia entre la pared y la ranura.

```python
## M. Osorio Mirambell - mosorio@fing.edu.uy
## 202108 - v1.0

import numpy as np
import matplotlib.pyplot as plt

################# BLOQUE A MODIFICAR #################
# Ancho de la rendija (en um)
b = 400;
# Longitud de onda de la radiación electromagnética (en nm)
wl = 650;
# Distancia entra la pared y la ranura (en cm)
D = 50;
######################################################

# Conversión de todas las dimensiones a metros
b = b*1.E-06; wl = wl*1.E-09; D = D*1.E-02;

# Definición de vector de ángulos, considerando que tan(ang)=x/D (x = distancia al punto de medición; D=distancia a la pantalla)
ang = np.linspace(-np.pi/200,np.pi/200,2000);
# Conversión: x = D*tan(ang) + pasaje a cm
dX = np.tan(ang)*np.array(D)*100;
# Brillo observado en la pantalla
intensity = (np.sin(np.pi*b*np.tan(ang)/wl)/(np.pi*b*np.tan(ang)/wl))**2;

plt.plot(dX,intensity,'r-')
plt.xlabel('distancia al centro [cm]')
plt.ylabel('intensidad observada [AU]')
plt.grid(True)
plt.show()
```

## Difracción por una red

Una red de difracción es un conjunto de elementos difractores equiespaciados. Existen redes por reflexión y por transmisión de la luz. Las que se usarán en la curso transmiten y reflejan al mismo tiempo.

Para analizar la difracción de una red, se considerará el caso de la difracción producida por un conjunto de rendijas de igual ancho *b* situadas todas en un mismo plano y espaciadas regularmente una distancia *a*. El patrón de la luz transmitida por la red está caracterizado por dos términos: la intensidad de la difracción producida por una rendija de ancho *b*, y la intensidad debida a la interferencia de *N* rendijas (de ancho *b*) separadas una distancia *a*.

Como efecto de la interferencia de las ondas difractadas, se pueden observar *m* diferentes órdenes de ondas que se mueven desde la red, como se muestra a continuación:
<img src="./imgDiff/redfrente.jpg" alt="Drawing" style="width: 600px;"/>

La expresión de la intensidad de luz medida del patrón de difracción es:

\begin{equation}
I(\theta)=\frac{I_0}{N^2}\left(\frac{\sin(\beta)}{\beta}\right)^2\left(\frac{\sin(N \alpha)}{\alpha}\right)^2
\end{equation}

donde *N* es el número de rendijas simples, $\beta$ la diferencia de fase entre las ondas difractadas por una rendija simple y $\alpha$ es la diferencia de fase entre las ondas difractadas por las *N* rendijas. A continuación se muestran  la distribución de intensidad y la foto del patrón generado hasta el orden 1. El patrón de la luz difractada consistirá en una serie de franjas brillantes, correspondientes a los máximos principales de la interferencia de *N* fuentes.

El siguiente script permite simular la intensidad observada, variando el ancho de la rendija, la longitud de onda, la distancia entre la pared y la ranura, el número de ranuras y la distancias entre sí. 

```python
## M. Osorio Mirambell - mosorio@fing.edu.uy
## 202108 - v1.0

import numpy as np
import matplotlib.pyplot as plt

################# BLOQUE A MODIFICAR #################
# Ancho de la rendija (en um)
b = 250;
# Longitud de onda de la radiación electromagnética (en nm)
wl = 632;
# Distancia entra la pared y la ranura (en cm)
D = 50;
# Número de ranuras
N = 20
# Distancia entre ranuras (en mm)
a = 1;
######################################################

# Conversión de todas las dimensiones a metros
b = b*1.E-06; wl = wl*1.E-09; D = D*1.E-02; a = a*1.E-03;

# Definición de vector de ángulos, considerando que tan(ang)=x/D (x = distancia al punto de medición; D=distancia a la pantalla)
ang = np.linspace(-np.pi/200,np.pi/200,2000);
# Conversión: x = D*tan(ang) + pasaje a cm
dX = np.tan(ang)*np.array(D)*100;
# Brillo observado en la pantalla
intensity = (np.sin(np.pi*b*np.tan(ang)/wl)/(np.pi*b*np.tan(ang)/wl))**2 * (np.sin(N*np.pi*a*np.tan(ang)/wl)/(N*np.sin(np.pi*a*np.tan(ang)/wl)))**2
# Corrección por problema en el centro del vector (intensidad máxima)
intensity[round(np.size(intensity)/2)]=1
# Envolvente de la curva
env = (np.sin(np.pi*b*np.tan(ang)/wl)/(np.pi*b*np.tan(ang)/wl))**2;

plt.plot(dX,intensity)
plt.plot(dX,env,'r--')
plt.xlabel('distancia al centro [cm]')
plt.ylabel('intensidad observada [AU]')
plt.grid(True)
plt.show()
```
